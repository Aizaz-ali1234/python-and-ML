#### Attribute
Definition: A value or property that belongs to an object.

Access: You just read it — no parentheses ().

Purpose: Stores information about the object.

```python
import pandas as pd

df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})

print(df.shape)   # (3, 2) → attribute (no parentheses)
print(df.columns) # Index(['A', 'B'], dtype='object')
```

2. Method
Definition: A function that belongs to an object.

Access: You call it — with parentheses ().

Purpose: Performs an action on the object, possibly changing it or returning something new.

```python
print(df.head())      # Shows first 5 rows → method
print(df.describe())  # Calculates statistics → method
```

#### How do I filter rows of a pandas DataFrame by column value?

```python
import pandas as pd
```

```python
movies=pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
type(True)
```

```python
booliens=[]
for length in movies.duration:
    if length>=180:
        booliens.append(True)
    else:
        booliens.append(False)
```

```python
booliens[0:5]
```

now we convert the booliens to pandas series.

```python
is_long= pd.Series(booliens)
```

```python
is_long.head()
```

```python
movies[is_long]
```

we can use the direct mehtod without for loop.

```python
is_long=movies.duration>=200
is_long.head()
```

```python
movies[movies.duration>=200]
```

movies.duration >= 200

This creates a Boolean Series — True for rows where duration is 200 or more, False otherwise.

This is Boolean indexing.

When you pass a Boolean Series inside movies[...], pandas returns only the rows where the value is True.

Why movies appears twice?

The first movies is the DataFrame you’re filtering.

The second movies (inside the condition) is to access the duration column from that DataFrame.

#### How do I apply multiple filter criteria to a pandas DataFrame?

```python
import pandas as pd
```

```python
movies= pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
True and True
```

```python
True & False
```

you need to write each situation in paranthesis and separator with & or| .

```python
movies[(movies.duration>=200) & (movies.genre=='Drama')]
```

```python
movies[(movies.duration>=200) | (movies.genre=='Drama')]
```

```python
(movies.duration>=200) | (movies.genre=='Drama')
```

```python
(movies.duration>=200) & (movies.genre=='Drama')
```

```python
movies[movies.genre.isin(['Crime', 'Action','Drama'])]# this two are the same identical thing.or 
movies[(movies.genre == 'Crime')|(movies.genre== 'Drama')|(movies.genre=='Action')]
```
