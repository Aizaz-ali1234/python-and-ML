### Dictionary Comprehension

Create a dictionary in python we use a dictionar comprehension in pythonjust like list comprehensions, but it builds key-vaule pairs instead.

```python
number={number:number*number for number in range(10)}
print(number)
```

#### Convert list to dictionary

```python
names = ['Ali', 'Aizaz', 'Rahman']
marks = [90, 85, 95]

student_marks = {name: mark for name, mark in zip(names, marks)}
print(student_marks)
```

#### With Condition (If Filter)

```python
{x:x**2 for x in range(10) if x%2==0}
```

```python
{x:("even" if x%2==0 else "odd") for x in range(6)}
```

```python
original = {'a': 1, 'b': 2, 'c': 3}
swapped = {value: key for key, value in original.items()}
print(swapped)
```

| Use Case                    | Example                               |
| --------------------------- | ------------------------------------- |
| Creating a dictionary       | `{x: x**2 for x in range(5)}`         |
| Filtering items             | `{x: x for x in range(10) if x%2==0}` |
| Transforming keys or values | `{k.upper(): v for k,v in d.items()}` |
| Inverting dictionary        | `{v: k for k, v in d.items()}`        |
                                                              

```python
power_of_two={}
for integer in range(1,10):
    power_of_two[integer]=2**integer
print(power_of_two)
```

```python
power_of_two={integer:2**integer for integer in range(1,10)}
print(power_of_two)
```

```python
fruits = {
    "apple": 100.00,
    "banana": 20.50,
    "cherry": 200.00,
}

{fruite:round(price *.95,4) for fruite,price in fruits.items()}
# if you need to decrease the price of fruite in a dictionary
```
