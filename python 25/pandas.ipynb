#### Attribute
Definition: A value or property that belongs to an object.

Access: You just read it — no parentheses ().

Purpose: Stores information about the object.

```python
import pandas as pd

df = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})

print(df.shape)   # (3, 2) → attribute (no parentheses)
print(df.columns) # Index(['A', 'B'], dtype='object')
```

2. Method
Definition: A function that belongs to an object.

Access: You call it — with parentheses ().

Purpose: Performs an action on the object, possibly changing it or returning something new.

```python
print(df.head())      # Shows first 5 rows → method
print(df.describe())  # Calculates statistics → method
```

#### How do I create a dummy Variable in Pandas?
In Pandas, a dummy variable means exactly the statistical definition — it’s a column of 0s and 1s used to represent categorical data in numeric form.

Why we need dummy variables in Pandas?
Machine learning models and many statistical methods can’t directly work with text labels (like "Red", "Blue", "Green"), so we convert them into numeric columns — one for each category.

```python
import pandas as pd
```

```python
train= pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
train['sex_male'] = train.Sex.map({'female': 0, 'male': 1})# the .map() function is used to transform each element of a Series one-by-one
train.head()
```

```python
train.Embarked.value_counts()
```

In the Titanic dataset, the Embarked column tells you the port where each passenger boarded the Titanic.

It has three main possible values:
| Code | Port Name   | Location     |
| ---- | ----------- | ------------ |
| `C`  | Cherbourg   | France       |
| `Q`  | Queenstown  | Ireland      |
| `S`  | Southampton | England (UK) |


```python
embarked_dummies=pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:,1:]
```

```python
train=pd.concat([train, embarked_dummies],axis=1)
train.head()
```
